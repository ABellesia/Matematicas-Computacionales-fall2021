```python
import matplotlib.pyplot as plt
```

```python
class Sierpinski:
    def __init__(self, x1, y1, x2, y2, x3, y3):
        self.x1 = x1
        self.y1 = y1
        self.x2 = x2
        self.y2 = y2
        self.x3 = x3
        self.y3 = y3
    
    def _generar(self, x1, y1, x2, y2, x3, y3, iteracion):
        
        if iteracion == 0:
            self.dibujar(x1, y1, x2, y2, x3, y3)
            
        elif iteracion > 0:
            
            xi=x1
            yi=y1
            
            xj=x2
            yj=y2
            
            xk=x3
            yk=y3
            
            x1 = (xi+xj)/2
            y1 = (yi+yj)/2
            
            x2 = (xj+xk)/2
            y2 = (yj+yk)/2
            
            x3 = (xk+xi)/2
            y3 = (yk+yi)/2

            self._generar(xi, yi, x1, y1, x3, y3, iteracion - 1)
            self._generar(x1, y1, x2, y2, xj, yj, iteracion - 1)
            self._generar(x2, y2, x3, y3, xk, yk, iteracion - 1)
            
    def dibujar(self, x1, y1, x2, y2, x3, y3):
        plt.plot([x1, x2, x3, x1], [y1,y2, y3, y1], lw=1.5, color='black')
        
    def generar(self, iteraciones=0):
        return self._generar(self.x1, self.y1, self.x2, self.y2, self.x3, self.y3, iteraciones)
```

```python
triangulito=Sierpinski(0,0,1,0,1/2,1)
```

```python
triangulito.generar(4)
```
