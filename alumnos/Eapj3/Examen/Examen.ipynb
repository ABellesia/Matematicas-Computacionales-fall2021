### Problema 2

El [**atractor de Rössler**](http://en.wikipedia.org/wiki/R%C3%B6ssler_attractor) esta descrito por el siguiente conjunto de ecuaciones:

$$
\frac{dx}{dt} = -y -z
$$

$$
\frac{dy}{dt} = x + ay
$$

$$
\frac{dz}{dt} = b + z(x-c)
$$

(a) Resuelva las ecuaciones numéricamente para 

$$
a=0.13 \quad b=0.2 \quad c=6.5 
$$

y condiciones iniciales

$$
x(0)=0 \quad y(0)=0 \quad z(0)=0 
$$

use el método de Runge-Kutta de 2do orden.

<font size="3.5">Hola. Tras algunos intentos fallidos con el Runge-Kutta, que francamente no entendí muy bien ni a papel y lápiz, decidí por lo menos dar una solución usando el método de Euler, que logré captar un poco mejor :s</font>

```python
vals = np.zeros([20000,3])
```

```python
for i in range(19999):
    vals[i+1]= vals[i]+0.01*np.array([-vals[i][1]-vals[i][2],vals[i][0]+0.13*vals[i][1],0.2+vals[i][2]*(vals[i][0]-6.5)])
```

```python
t = np.linspace(0,199.99,20000)
```

(b) Muestra en una gráfica el comportamiento de las soluciones en el tiempo (i.e. grafica $x(t)$, $y(t)$ y $z(t)$).

```python
plt.plot(t,vals[:,0])
plt.title("x(t)")
plt.show()
```

```python
plt.plot(t,vals[:,1])
plt.title("y(t)")
plt.show()
```

```python
plt.plot(t,vals[:,2])
plt.title("z(t)")
plt.show()
```
