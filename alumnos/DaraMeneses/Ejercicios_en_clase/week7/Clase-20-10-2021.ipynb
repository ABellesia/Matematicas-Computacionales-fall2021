```r
flights %>%
    mutate(
    speed = distance / air_time * 60,
    speed2 = speed*2
    )
```

```r
flights %>%
    group_by(month, day) %>%
    summarise(delay = mean(dep_delay, na.rm = TRUE)) %>%
    arrange(desc(delay))
```
