### Problema 1

(a) Usando `Simpy`, declara las funciones:
$$
y(x) = \cos(x)
$$

$$
z(x) = \sin(x)
$$

$$
w(x) = \frac{1}{\sin(x) + \cos(2x)}
$$



```python
# Se declaran las variables junto con las funciones
y=Function('y') 
z=Function("z")
w=Function("w")
g=Function("g")
x, x0=symbols("x, x_0")
```

```python
y_x = Eq(y(x), cos(x))
y_x
```

```python
z_x = Eq(z(x), sin(x))
z_x
```

```python
w_x = Eq(w(x), 1/(sin(x)+cos(2*x)))
w_x
```

(b) Obtén la derivada de $g(x)$

$$
g(x) = y(x)*z(x)
$$



```python
g_x = Eq(g(x), y_x.rhs*z_x.rhs)
g_x
```

```python
Eq(Derivative(g_x.lhs),simplify(diff(g_x.rhs)))
```

(c) Grafica $w(x)$ en el rango $[0,1]$


```python
f_w = lambda x: 1/(sin(x)+cos(2*x)) #hacemos una función para obtener valores de y

valores_x = np.linspace(0,1)
valores_y =[f_w(i) for i in valores_x]
```

```python
plt.figure(figsize=(6, 6), dpi=100)
plt.plot(valores_x, valores_y,linestyle="-", label = 'y=w(x)')
plt.legend()
plt.xlabel('X', fontsize=15)
plt.ylabel('Y', fontsize=15)
plt.title("Gráfica w(x)", fontsize=30)
```


(d) Integra de manera indefinida $g(x)$ y luego evalúala desde $0$ a $1$.


```python
# AQUI LA INTEGRAL DE g(x) D:
int_g= Eq(integrate(g_x.lhs), integrate(g_x.rhs))
int_g
```

```python
Eq(integrate(g_x.lhs, (x,0,1)),int_g.rhs.subs(x,1)-int_g.rhs.subs(x,0)) #Por el T.F.C. 
```


(d) Integra de manera indefinida $w(x)$ y luego evalúala desde $0$ a $1$. (Respuesta ORIGINAL)



```python
int_w= Eq(integrate(w_x.lhs), integrate(w_x.rhs))
int_w
```

```python
# integramos la serie de taylor de w(x) alrededor de 1/2 de orden 50,
#ya que el metodo de sympy integrate no da una respuesta
# esto nos debe dar una aproximacion muy cercana para obtener el valor de la integral evaluada de 0 a 1.
int_w_s= Eq(integrate(w_x.lhs), integrate(w_x.rhs.series(x,.5,51).removeO()))
int_w_s
```

```python
eval_int_w_s=int_w_s.rhs.subs(x,1)-int_w_s.rhs.subs(x,0)#Por el T.F.C. 
Eq(integrate(w_x.lhs, (x,0,1)),round(eval_int_w_s,5)) 
```

### Problema 2

El [**atractor de Rössler**](http://en.wikipedia.org/wiki/R%C3%B6ssler_attractor) esta descrito por el siguiente conjunto de ecuaciones:

$$
\frac{dx}{dt} = -y -z
$$

$$
\frac{dy}{dt} = x + ay
$$

$$
\frac{dz}{dt} = b + z(x-c)
$$

```python
def RK2(y, t, dt, derivadas): # copiado de la clase D:
    k0 = dt*derivadas(y, t)
    k1 = dt*derivadas(y + k0, t + dt)
    y_next = y + 0.5*(k0 + k1)
    
    return y_next
```

(a) Resuelva las ecuaciones numéricamente para 

$$
a=0.13 \quad b=0.2 \quad c=6.5 
$$

y condiciones iniciales

$$
x(0)=0 \quad y(0)=0 \quad z(0)=0 
$$

use el método de Runge-Kutta de 2do orden.

El método de Runge-Kutta nos dice que las soluciones son 
$$
x_{i+1} = x_i + \ (-y-z)\Delta t
$$$$ 
y_{i+1} = y_i + \ (x + ay)  \Delta t
$$$$ 
z_{i+1} = z_i + \ (b + z(x-c))  \Delta t
$$


```python
def atractor_rossler(y, t): #asi describimos es sistema del atractor_rossler, lo que va en "derivadas" de RK2
    a0= -y[1]-y[2]
    a1= y[0] + .13*y[1]
    a2= .2 + y[2]*(y[0]-6.5)
    return np.array([a0,a1,a2])
```

```python
N = 100000 # número de pasos
t = np.linspace(0,100,N) # 100 es arbitrario, tiempo
dt = 100/float(N-1) # tamaño del paso
```

```python
y = np.zeros([N,3]) #representa que los valores iniciales son cero y cada columna representa a x,y,z respectivamente
```

```python
# así se calculan los siguientes x,y,z
for i in range(N-1):
    y[i+1] = RK2(y[i], t[i], dt, atractor_rossler)
y #valores expresados como [x,y,z]
```

(b) Muestra en una gráfica el comportamiento de las soluciones en el tiempo (i.e. grafica $x(t)$, $y(t)$ y $z(t)$).

```python
atx = [y[i,0] for i in range(N)]
aty = [y[i,1] for i in range(N)]
atz = [y[i,2] for i in range(N)]
```

```python
plt.figure(figsize=(6, 6), dpi=100)
plt.title("Atractor de Rössler", fontsize=30)
plt.xlabel('Tiempo', fontsize=15)
plt.ylabel('x,y,z', fontsize=15)
plt.plot(t, atx, linewidth=2,label = "x(t)")
plt.plot(t, aty, linewidth=2,color="red",label = "y(t)")
plt.plot(t, atz, linewidth=2,color="green",label = "z(t)")
plt.legend()
```

(c) Muestra como se ve el **atractor de Rössler** en 3D (i.e. en el espacio).

```python
fig = plt.figure(figsize=(9, 5), dpi=100)
ax = fig.add_subplot(1,1,1, projection='3d')
ax.plot(atx, aty, atz, label="Soluciones")
plt.title("Atractor de Rössler", fontsize=30)
plt.legend(loc='best')
ax.set_xlabel("x", fontsize=15)
ax.set_ylabel("y", fontsize=15)
ax.set_zlabel("z", fontsize=15)
```
